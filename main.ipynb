```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
dfinput[~dfinput['total_sqft'].apply(is_float)]
```

```python
df2=dfinput.copy()
```

```python
df2['price']=dfout.copy()
```

```python
def sqftnum(x):
    t = x.split('-')
    if len(t) == 2:
        return (float(t[0]) + float(t[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df2['sqft'] =  df2['total_sqft'].apply(sqftnum)
```

```python
df2
```

```python
df2 = df2[df2.sqft.notnull()]
```

```python
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
scores = []
```

```python
algos = {
    'Linear Regression' : {
        'model' : LinearRegression(),
        'params' : {
            'n_jobs' : [1,2,3]
        }
    },
    'Lasso' : {
        'model': Lasso(),
        'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']}
    },
    'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
}
```

```python
for algo, config in algos.items():
    gs = GridSearchCV(config['model'],config['params'], cv = cv)
    gs.fit(X,Y)
    scores.append({
        'model': algo,
        'best testscore' : gs.best_score_, 
        'best params' : gs.best_params_
    })
```

```python
scores
```
